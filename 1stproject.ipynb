```python
df_list = []

# Your loop was using `range(1,)` which means it runs 0 times. Use `range(1)` if you want to run once.
for i in range(1):
    # The URL does not need formatting with `i`, so we remove `.format(i)`
    response = requests.get('https://raw.githubusercontent.com/sondosaabed/Palestinian-Movies-JSON-Dataset/main/palestinian_movies.json')
    temp_df = pd.DataFrame(response.json()['movies'])
    df_list.append(temp_df)

# Combine all collected DataFrames
df = pd.concat(df_list, ignore_index=True)

# Show the first few rows
print(df.head())
```
